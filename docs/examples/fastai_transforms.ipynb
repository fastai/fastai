```python
datadir = Path('./data/catdog')
```

```python
test_cat_fname = 'cat.10013.jpg'
```

```python
imgarr =  np.array(PIL.Image.open(os.path.join(datadir, f'train/cat/{test_cat_fname}')))
```

```python
plt.imshow(imgarr)
plt.show()
```

```python
tmp_arr = zoom_cv(imgarr, 2)
plt.imshow(tmp_arr)
plt.show()
```

```python
tmp_arr = stretch_cv(imgarr, 1,3)
plt.imshow(tmp_arr)
plt.show()
```

```python
tmp_arr = dihedral(imgarr, 7)
plt.imshow(tmp_arr)
plt.show()
```

```python
tmp_arr = lighting(imgarr, 0, 50)
print(np.average(imgarr))
plt.imshow(tmp_arr)
plt.show()
```

```python
tmp_arr = cutout(imgarr, 10, 20)
print(np.average(imgarr))
plt.imshow(tmp_arr)
plt.show()
```

```python
tmp_arr = scale_to(imgarr,3, 10)
print(np.average(imgarr))
plt.imshow(tmp_arr)
plt.show()
```

```python
tmp_arr = crop(imgarr,0, 30, 400)
plt.imshow(tmp_arr)
plt.show()
```

```python
m = np.mean(imgarr)
s = np.std(imgarr)
print(m, s)
normalizer = Normalize(174, 60)
tmp_arr, _ = normalizer(imgarr)
plt.imshow(tmp_arr)
plt.show()
```
